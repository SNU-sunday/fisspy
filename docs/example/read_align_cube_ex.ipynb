# Read Aligned Cube Data

If you don't have any aligned data, please see the [Align Data](/fisspy/align/data/) first. After aligning the series of data, you can easily draw the interactive figure. It use the [makeTDmap](/fisspy/analysis/tdmap) class to draw the figure. Note that you should change the matplotlib backend to one of interactive backend.

```python
import fisspy
import matplotlib.pyplot as plt

ac = fisspy.read.AlignCube(r'D:\Data\140603\140603_alosv_A.npz')
ac.imshow(cmap=plt.cm.RdBu_r, clim=[-5,5])
```

Interactive key are described in the example of the [makeTDmap example](/fisspy/analysis/tdmap/example).
